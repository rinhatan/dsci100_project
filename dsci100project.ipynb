**DATA DESCRIPTION**

```r
player_data <- "https://raw.githubusercontent.com/rinhatan/dsci100_project/refs/heads/main/players.csv"
players <- read_csv(player_data)
players
```

There are 196 observations and 7 variables. The data frame gives insight to how the player felt about the game, how invested they were and provides information about the players.

**Player data frame variables**:

  - `experience`: Level of experience the player has with games (Character)
  - `subscribe`: Whether they subscribed to a game-related newsletter (Logical)
  - `hashedEmail`: Player ID (Character)
  - `played_hours`: Hours the game was played (Double)
  - `name`: Name of player (Character)
  - `gender`: Gender of player (Character)
  - `age`: Age of player (Double)

Since `played_hours` is in hours, those who only contributed only minutes are not visible in the data frame as the numbers only goes to the tenths. This would have been avoided by providing minutes or seconds as time indicators.

```r
sessions_data <- "https://raw.githubusercontent.com/rinhatan/dsci100_project/refs/heads/main/sessions.csv"
sessions <- read_csv(sessions_data)
sessions
```

There are 1535 observations and 5 variables. The session data frame gives insight to how long the session was managed in addition to when the game was played by who given the `hashedEmail` variable.

**Session data frame:**

  - `hashedEmail`: Player ID (Character)
  - `start_time`: The date game started (Character)
  - `end_time`: The date game ended (Character)
  - `original_start_time`: UNIX start time (Double)
  - `original_end_time`: UNIX end time (Double)

The UNIX time may not be useful as it is unable represent a precise tracking time and the values are large. Additionally, since the `start_time` and `end_time` are not actually set as a date and is a character type, therefore will be unable to calculate date based questions.

```r
played_hours_vs_age <- ggplot(players, aes(x = Age)) +
                geom_bar() +
                xlab("Age of player") +
                ylab("Count of participants") + 
                ggtitle("The age of participants") + 
                theme(text = element_text(size = 12))
played_hours_vs_age
```

This graph suggests that the most common age of those who played MineCraft was 17. We can also see that the youngest participant was 8 and the oldest participant was 50.

```r
age_vs_played_hours <- ggplot(players, aes(x = Age, y = played_hours)) +
                geom_point() +
                xlab("Age of participants") +
                ylab("Hours of game played") +
                ggtitle("Age vs. time of game played") +
                theme(text = element_text(size = 12))
age_vs_played_hours
```

In the graph, majority of participants had only played the game for a small portion as many points are close to 0. We can also see that there is no strong relationship between the points nor a linear relationship.

**METHODS AND PLAN**

Since the classification method uses quantitative variables to predict the value of another categorical variable, the question asked fits the use of the classification method the most. In this case, we are assuming that if the data points are closer to each other, the more similar they are when performing a prediction. A weakness to the classification model is how with a large training set, the information becomes slower to generate. Additonally, it may not perform well if the classes are imbalanced, though we could overcome by oversampling the rare class. Furthermore, standardizing the `played_hours` and `Age` is important, as well as separating the data between the amateurs and pros for finalizing the difference between the two explanatory variables.
