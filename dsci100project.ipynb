**DATA DESCRIPTION**

```r
player_data <- "https://raw.githubusercontent.com/rinhatan/dsci100_project/refs/heads/main/players.csv"
players <- read_csv(player_data)
players
```

In the player data frame, we are able to observe 196 observations and 7 variables. The player data frame gives insight to how the player felt about the game, how invested they were as well as providing information on the player themselves to see if those variables influence whether or not they enjoyed playing MineCraft or not.

**Player data frame variables**:

  - `experience`: The amount of experience the player has with games being seperated into either Amateur, Regular, Veteran, or Pro. Currently with the character data type.
  - `subscribe`: Whether the player had decided to subscribe to a game-related newsletter. Currently with the logical data type.
  - `hashedEmail`: Player ID given to the player for the game. Currently with the character data type.
  - `played_hours`: Number of hours the game was played by the player. Currently with the double data type.
  - `name`: Name of the player. Currently with the character data type.
  - `gender`: Gender of the player. Currently with the character data type.
  - `age`: The age of the player. Currently with the double data type.

In the original player data frame, I believe that the `experience` and `gender` variables should be in a factor data type for the data be easier to work with in the future. Furthermore, creating another column with `played_minutes` additional to `played_hours` using the mutate function may be beneficial as some players had not played enough hours for the numbers to be easy to identify how many hours they played. By giving `played_minutes`, it will be able display a much more detailed and descriptive piece of information than hours. Furthermore, since the `played_hours` variable is counted only in hours, those who only contributed only small amount of time (for example, only 5 minutes) cannot be visible in the data frame as the numbers only goes to the tenths and not the hundredths. This would have been better avoided by providing a `played_minutes` or `played_seconds` with minutes or seconds as the time indicators.

```r
played_hours_vs_age <- ggplot(players, aes(x = Age)) +
                geom_bar() +
                xlab("Age of player") +
                ylab("Count of participants") + 
                ggtitle("The age of participants") + 
                theme(text = element_text(size = 12))
played_hours_vs_age
```

This graph suggests that the most common age of those who played MineCraft was 17. We can also see that the youngest participant was 8 years old and the oldest participant was 50 years old.

```r
subscribe_vs_gender <- ggplot(players, aes(x = gender, fill = subscribe)) +
                geom_bar() +
                xlab("Gender of participants") +
                ylab("Number of participants") +
                labs(fill = "Subscribed?") + 
                ggtitle("Gender vs. subscription") +
                theme(text = element_text(size = 12))
subscribe_vs_gender
```

This graph displays the distribution between gender and whether or not they decided to subscribe or not. In this, we are able to observe that men is the most common gender who participated in this experiment, the next being women. We see that all except those who picked "Prefer not to say" has a high ratio of subscribing to the game news, though that is not enough to clearly see what gender most enjoyed the game. Since the number of participants per gender is different, it is diffucult to compare ratios as things may change with more trials with more people.

```r
age_vs_played_hours <- ggplot(players, aes(x = Age, y = played_hours)) +
                geom_point() +
                xlab("Age of participants") +
                ylab("Hours of game played") +
                ggtitle("Age vs. time of game played") +
                theme(text = element_text(size = 12))
age_vs_played_hours
```

This graph roughly displays the number of hours played by the participant with their age. We are able to observe that a majority of participants had only played the game for a small portion as many points are closer to 0 in the y-axis. Additionally, we can see that there is no strong relationship between the points nor is there a linear relationship.

**METHODS AND PLAN**

Since the classification method uses one or more quantitative variables to predict the value of another categorical variable, the question being asked fits the use of the classification method the most (since we are using `played_hours` and `Age`, two quantitative variables to predict `subscribe`, a categorical variable). In this case, we are assuming that if the data points are closer to each other, the more similar they are when performing a prediction, though are going to check if the assumption is right using to determine whether the two variables have high accuracy, recall, and precision. A weakness to the classification model is how with a large training set, the information may become slower to access. Additonally, it may not perform well if the classes are imbalanced, in which we could avoid by oversampling the rare class. Furthermore, standardizing the `played_hours` and `Age` is a key part of processing the data frame as well as seperating the data between the amateurs and pros to finalize whether there is a difference in the two explanatory variables from their experience.

Since typically the training set is between 50% to 95% of the data while the test set is the remaining 5% to 50%, I believe using 75% of the data for training and the remaining 25% of the data for testing is a reasonable proportion. Splitting will be done after we standardize the whole data frame and make sure the predicting variable is a factor instead of a logical data type. Lastly, we will use 5-fold cross-validation to achieve a better choice of the K value and higher accuracy.
